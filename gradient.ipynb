```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to minimize (example: quadratic function)
def loss_function(x):
    return x**3

# Define the derivative of the function
def gradient(x):
    return 2*x + 5

def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    path = [(x, loss_function(x))]  # Store the path for visualization

    for i in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        path.append((x, loss_function(x)))

    return x, path

# Hyperparameters
initial_x = 0.0
learning_rate = 0.1
num_iterations = 100

# Run gradient descent
final_x, path = gradient_descent(initial_x, learning_rate, num_iterations)

# Print final results
print("Gradient Descent Finished")
print(f"Final x = {final_x}, Final Loss = {loss_function(final_x)}")

# Extract x and loss values for visualization
x_values, loss_values = zip(*path)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, loss_values, '-o', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('Loss')
plt.title('Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
